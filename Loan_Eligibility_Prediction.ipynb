### **Splitting the dataset**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
x=df.drop(columns=['Loan_ID','Loan_Status'])
y=df['Loan_Status']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
Y_train
```

```python
X_test
```

```python
Y_train
```

## **Logistic Regression**

```python
logisticreg = LogisticRegression()
logisticreg.fit(X_train,Y_train)
Y_predict=logisticreg.predict(X_test)
result=pd.DataFrame({'Actual':Y_test,'Predicted':Y_predict})
print(result)
print("accuracy of logistic regression:",metrics.accuracy_score(Y_test,Y_predict))
```

## **Support Vector Machine**

```python
Svmc = SVC(kernel='linear')
Svmc.fit(X_train,Y_train)
Y_predict = Svmc.predict(X_test)
result=pd.DataFrame({'Actual':Y_test,'Predicted':Y_predict})
print(result)
print("Accuracy of Support vector machine classifier:",metrics.accuracy_score(Y_test,Y_predict))
```

```python

print(confusion_matrix(Y_test,Y_predict))
```

```python
print(classification_report(Y_test, Y_predict))
```

## **Random Forest**

```python
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
Y_predict = rfc.predict(X_test)
print("Accuracy of Random forest:",metrics.accuracy_score(Y_test,Y_predict))
```
